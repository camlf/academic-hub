# Data view tutorial
--------------

Data view is a data display tool to map data into a customizable table format. Data view has enhanced the people's ability to organize, present, and interpolate the data taken from multiple sensors and sources. We utilize Data view and store the Data view data in the OSIsoft Cloud Services(OCS). After requesting and reading the data view, its result is presented in a data frame in the desired shape.

The dataset used below is from the Deschutes Brewery, with data coming from fermentor vessels.
This notebook shows the steps involved in creating and using Dataviews, and thus students become familiar with the concept of Data view.
The primary purpose of this data view tutorial is to inform students the steps involved in using Data view. For this example, we will use the 2-hours worth of the bottom temperature sensor data from fermenter vessel 31 starting from 00:00 on March 17, 2017 to 02:00 on March 17, 2017.  



## The overview of the tutorial:

__Part 1. Data retrieve without using a data view__
 
 Request the stream data of the fermentation vessel 31([1a](#section_1a)), iterate through the stream data matching the set criteria and extract its value([1b](#section_1b)), and present the data in the data frame ([1c](#section_1c)).

__Part 2. Data retrieve using a data view__ 

 Find the dataview Ids pre-installed in OCS([2a](#section_2a)), and request and read the data of fermentation vessel 31 for the given time frame([2b](#section_2b)).
 
__Part 3. Data retrieve for multiple fermentation vessels (31 to 36) using the data views__

Make a list of data view ids that correspond to the fermenters (31 to 36)([3a](#section_3a)), and read the multiple data views corresponding to the fermenters 31 to 36 for the given time frame ([3b](#section_3b)).

## Data Architecture
-------------------
    
Sequential data store(SDS) is the overarching data architecture of OCS that governs the information flows, that presents the persistent data, and that controls where the data is stored and how it is used. The SDS architecture is represented by the SDS Tenant schematic shown below:
<br>

<img src = "https://apimgmtstelkv30lahnuj362.blob.core.windows.net/content/MediaLibrary/lehigh/images/tenant-namespace.png"  align="left">   

</br>   
where a tenant is a customer account divided into the Namespaces. Each namespace exist as a virtual separation across the namespaces and each store a set of multiple assets; for example, `fermenter_vessels` is our working namespace. Within the namespace, the streams exist to store the time-series data collecting the events (such as the data for the process parameters, or any types of time-series data). Data view, existing at the same level as streams, takes the stream data and display them in a desired and strcutured data frame. Data view is used for its versatile functionalities:  organizing all the data ((such as multiple sensors, calculations, and the settings) as rows of observations,  presenting the data in consistent shape, and interpolating between the data point.

## 1.0 Standard OCS initialization

````python
config = configparser.ConfigParser()
config.read("config.ini")

hub_client = HubClient(
    config.get("Access", "ApiVersion"),
    config.get("Access", "Tenant"),
    config.get("Access", "Resource"),
    config.get("Credentials", "ClientId"),
    config.get("Credentials", "ClientSecret"),
)

namespace_id = config.get("Configurations", "Namespace")
namespace_id
````

# Part 1. Data retrieve on OCS without using a data view

### 1a. Extract bottom temperture data for stream FV31
<a id="section_1a"></a>

````python
# retrieve stream data of fermentation vessel 31 on OSIsoft Cloud services(OCS)
fv31_streams = hub_client.Streams.getStreams(namespace_id, "fv31")
````

````python
# Defining the time frame of 2 hours
start_index = "2017-03-17T00:00"
end_index = "2017-03-17T02:00"
````

### 1b. Reading the bottom temperature data of fermentation vessel 31
<a id="section_1b"></a>
Iterate through each stream of fermentation vessel 31 that meets the specific criteria: stream description = `FV31 Bottom Temperature Control Value`. At the same time, extract the value of those stream data and make a list thereof. 

````python
streams_bottom_temp = []
for stream in fv31_streams:
    if stream.Description == "FV31 Bottom Temperature Control Value": 
        print(stream.Name, stream.Id, stream.TypeId, stream.Description)
        streamvalue = hub_client.Streams.getWindowValues(
            namespace_id, stream.Id, None, start_index, end_index
        )
        streams_bottom_temp.append(streamvalue)
````

### 1c. Presenting the bottom temperature data in the data frmae
<a id="section_1c"></a>
pass the bottom temperature stream data into the pandas data frame function

````python
# pass the list of the bottom temperature data and will return the value of the desired temperature
df_bottom_temp = pd.DataFrame(streams_bottom_temp[0])
# The return column name is shown as "Value" and is then renamed as "Bottom Temperature"
df_bottom_temp = df_bottom_temp.rename(columns={"Value": "Bottom Temperature"})
df_bottom_temp.head(10)
````
