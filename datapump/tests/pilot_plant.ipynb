```python
dv = hub.asset_dataviews(asset)
dv
```

```python
hub.dataview_definition(namespace_id, dv[0])
```

```python
df = hub.dataview_interpolated_pd(namespace_id, dv[0], "2020-10-14T00:00", "2020-11-30", "00:15:00")
df.to_csv("test.csv")
```

```python
mode_status = [c for c in df.columns if any(ss in c.lower() for ss in ["mode", "status"])]
mode_status
```

```python
print("--- Digital states found for each tag ---")
for c in mode_status:
    print(c, ": ", end="")
    print(", ".join(df.dropna()[c].unique()))
```
