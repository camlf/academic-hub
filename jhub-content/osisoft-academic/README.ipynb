# Welcome to the Academic JupyterHub!
#### Version 9-dev 

The Academic JupyterHub is OSIsoft private instance of [JupyterHub](https://zero-to-jupyterhub.readthedocs.io/en/stable/), a scalable, sustainable and easy to deploy multi-user server for Jupyter notebooks. It allows to deploy easily data science content to a wide audience. Instead of spending time installing required software or getting acquainted with a new virtual machine environement, students can right away start to learn, explore and experiment.  

After logging in with their credentials (right now need a Github account member of the osisoft-academic organization), a private environment is created for the student in a few minutes, including a permanent disk, with all the content needed to perform a live teaching lab or learn with self-service modules. Only a web browser is necessary to perform all the tasks. Additional content can be dynamically added through mechanism like [nbgitpuller](https://github.com/jupyterhub/nbgitpuller#what-is-it). 

**Find below a set of notebooks showing different features of the Academic JupyterHub. Clicking a link starts a notebook (blue link ending with .ipynb) with its own computing kernel (Python or R) including all required software pre-installed, ready to run. Execute notebook cells by typing Shift-Enter.**

## Notebook: [PI World 2018 Lab: Fit for Purpose - Layers of Analytics using the PI System – AF, MATLAB, Machine Learning](PIWorld2018/dryer.ipynb)

This is a R notebook to perform all the analysis and calculation of the first exercise of this lab, described below. 

![](images/dry-feed-pivision.png)

#### Exercise 1 – Alkylation Process Feed Dryer – Analytics 

In this lab, we use of PI Event Frames to label operating conditions captured during the regeneration process of two process feed dryers. The dryers are cycled back and forth to continuously remove moisture from the feed before it combines with a strong acid which serves as a catalyst for the chemical reaction. Even small amounts of moisture entering the reaction create a highly corrosive environment that will damage and compromise the equipment metallurgy. The modelling objective is to create a temperature profile representing proper regeneration of the dryer bed which is critical to this process.
